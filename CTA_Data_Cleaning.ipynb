```python
# For when "isArr" is added, that needs to be part of the criteria also

def clean_data(game):
    games = []
    for i in range(len(game)):
        if i == 0:
            game.loc[i][0]=re.sub(r"'",'',game.loc[i][0])
            game.loc[i][0]=re.sub(r'\[','',game.loc[i][0])

            game.loc[i][1]=re.sub(r"'",'',game.loc[i][1])
            game.loc[i][1]=re.sub(r' ','',game.loc[i][1])
            game.loc[i][2]=re.sub(r"'",'',game.loc[i][2])
            game.loc[i][2]=re.sub(r' ','',game.loc[i][2])
            game.loc[i][3]=re.sub(r"'",'',game.loc[i][3])
            game.loc[i][3]=re.sub(r' ','',game.loc[i][3])

            game.loc[i][4]=re.sub(r"'",'',game.loc[i][4])
            game.loc[i][4]=re.sub(r'\]','',game.loc[i][4])
            game.loc[i][4]=re.sub(r' 2021[0-9]{4} ','',game.loc[i][4])
        elif game.loc[i].rn == game.loc[i-1].rn and game.loc[i].station == game.loc[i-1].station and game.loc[i].isArr == game.loc[i-1].isArr and game.loc[i].direction == game.loc[i-1].direction:
            #print(i-1)
            games.append(i-1)
        
        game.loc[i][0]=re.sub(r"'",'',game.loc[i][0])
        game.loc[i][0]=re.sub(r'\[','',game.loc[i][0])
        
        game.loc[i][1]=re.sub(r"'",'',game.loc[i][1])
        game.loc[i][1]=re.sub(r' ','',game.loc[i][1])
        game.loc[i][2]=re.sub(r"'",'',game.loc[i][2])
        game.loc[i][2]=re.sub(r' ','',game.loc[i][2])
        game.loc[i][3]=re.sub(r"'",'',game.loc[i][3])
        game.loc[i][3]=re.sub(r' ','',game.loc[i][3])
        
        game.loc[i][4]=re.sub(r"'",'',game.loc[i][4])
        game.loc[i][4]=re.sub(r'\]','',game.loc[i][4])
        game.loc[i][4]=re.sub(r' 2021[0-9]{4} ','',game.loc[i][4])
        
        
    game = game.drop(games)
            
    return game
```

```python
def the_whole_deal(game):

    for i, r in enumerate(game.rn.unique()):

        gm = clean_data(game[game.rn == r].reset_index(drop=True)).reset_index(drop=True)
        gmr = run_funct(gm)
        runsdf = put_it_in_df(gmr)
        
    return runsdf.sort_values(by =['Merchandise Mart5']).reset_index(drop=True)
```
